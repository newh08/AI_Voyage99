## [MYCODE] 수능 국어 문제 풀기
- 최종 점수 58
- 문제 먼저 읽고 지문 읽기 전략, 정답으로 숫자만 출력

```python
import requests
from openai import OpenAI

# URL에서 JSON 데이터 가져오기
url = "https://raw.githubusercontent.com/NomaDamas/KICE_slayer_AI_Korean/refs/heads/master/data/2023_11_KICE.json"
response = requests.get(url)

# JSON 데이터를 Python 객체로 변환
data = response.json()

client = OpenAI(api_key="apikey")

temperature = 0.5  # 각 token을 샘플링할 때 사용하는 temperature 값입니다.
max_tokens = 4096  # 생성하는 최대 token 개수 입니다.
n = 1  # 같은 질의에 대해 몇 개의 답변을 출력할지 결정합니다.
frequency_penalty = 0.0  # 같은 단어가 반복적으로 나오는 것을 방지하기 위한 옵션입니다.

totalPoint = 0

for item in data:
    paragraph = item["paragraph"]

    for problem in item["problems"]:
        # 문제와 question_plus 결합
        question = problem["question"]
        if "question_plus" in problem:
            question += "\n" + problem["question_plus"]

        choices = "\n".join(
            [
                f"{idx} {choice}"
                for idx, choice in enumerate(problem["choices"], start=1)
            ]
        )
        answer = problem["answer"]
        score = problem["score"]

        prompt = (
            """역할:
당신은 수능 국어 문제를 정확히 분석하고 해결하는 전문가입니다.

목표:
주어진 지문과 문제를 읽고, 각 질문에 대해 꼼꼼히 검토하여 가장 적합한 정답을 오직 숫자(1~5) 형태로 출력합니다.

다음과 같은 형식으로 지문과 문제를 제공합니다:
- 문제: {문제 내용}
- 선택지: {선택지}
- 지문: {지문 내용}
"""
            + f"""

**문제 해결 방법 (문제를 먼저 읽는 전략):**  
1. 문제 유형 파악: 문제와 선택지를 먼저 읽어, 문제 유형(내용 일치/불일치, 빈칸 추론, 문장 배열, 어휘/어법, 주제/요지 파악 등)을 식별합니다.
2. 지문 분석: 지문을 꼼꼼히 읽어 주요 내용, 문맥, 논리 구조, 핵심 단어와 문장을 파악합니다.
3. 선택지 대조 및 검토: 각 선택지를 지문과 비교하여, 문제에서 요구하는 조건에 가장 적합하거나 부합하지 않는 선택지를 검토합니다.
4. 정답 결정: 논리적 추론과 비교를 통해 최적의 정답을 도출합니다.
5. 출력 형식: 반드시 숫자(1~5) 만 출력하며, 다른 기호나 텍스트는 포함하지 않습니다.

⚠️ **중요: 출력 형식: 반드시 숫자(1~5)만 출력합니다. 다른 문자나 기호(예: 마침표, 공백 등)를 포함하지 마세요.**
출력 예시:
- 정답이 1번이라면: 1
- 정답이 5번이라면: 5

문제 풀이 보조 전략:
- 내용 일치/불일치: 선택지의 진술이 지문 내용과 일치하는지 또는 불일치하는지 판단합니다.
- 빈칸 추론: 지문의 앞뒤 문맥과 논리를 분석하여 빈칸에 들어갈 적절한 선택지를 찾습니다.
- 문장 배열: 문장 간의 논리적 흐름과 연결어를 분석하여 가장 자연스러운 순서를 찾습니다.
- 어휘/어법: 지문의 어휘와 문법적 맥락을 분석하여 선택지와 비교합니다.
- 주제/요지 파악: 지문의 핵심 주제와 전체 흐름을 고려하여 가장 적절한 선택지를 선택합니다.

---
- 문제: {question}

- 선택지: {choices}

- 지문: {paragraph}"""
        )
        message = [{"role": "user", "content": prompt}]

        response = client.chat.completions.create(
            model="gpt-4o",
            messages=message,
            n=n,
            max_tokens=max_tokens,
            temperature=temperature,
            frequency_penalty=frequency_penalty,
        )
        GPT_answer = response.choices[0].message.content

        if int(GPT_answer) == answer:
            totalPoint += score
            print(totalPoint)

print("최종 점수: " + str(totalPoint))
```
